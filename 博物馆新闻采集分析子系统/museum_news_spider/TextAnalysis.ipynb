```python
# 从文件中读入停用词
stopwords_csv = pd.read_csv("F:/Codes/PySpider/demoSpider/jieba/stopwords.txt",
              index_col=False,
              quoting=3,
              sep="\t",
              names=['stopword'],
              encoding='utf-8') # quoting=3 全不引用

# 构建停用词列表
stopwords = []
for i in range(0, stopwords_csv.shape[0]):
    stopwords.append(stopwords_csv.iloc[i, 0])

# 构建情感词词典
word_dict = {}

# 从文件中读入情感词
with open("F:/Codes/PySpider/demoSpider/BosonNLP_sentiment_score/BosonNLP_sentiment_score.txt", mode='r', encoding='UTF-8') as f:
    lines = f.readlines()
    for line in lines: 
        index = line.find(" ")
        word = line[0:index]
        score = line[index+1:]
        if not score.isspace():
            if score[0] == '-':
                score = - float(score[1:])
            else:
                score = float(score)
            word_dict[word] = score
```

```python
stopwords
```

```python
word_dict
```

```python
# 根据输入的单词得到情感词典中的分数
def getScore(word, word_dict):
    tf_idf = word[1]
    word = word[0]
    return tf_idf * word_dict.get(word, 0)
```

```python
for i in range(len(text)):
    word_list = text.iloc[i, :]['content']

    # 分词后去掉语气词、空格和数字
    l1 = []
    for word in word_list:
        if not word.isdigit() and not word.isspace() and word not in stopwords :
            l1.append(word)

    # 提取出TF-IDF的语气词频
    tf_idf_keyword_list = analyse.extract_tags(word_list, topK=50, withWeight=True, allowPOS=())

    # 去掉空语气词并保留其语气词词频，并计算其情感得分
    l2 = []
    tf_idf_score = 0
    for word in tf_idf_keyword_list:
        if not word[0].isdigit() and not word[0].isspace() and word[0] not in stopwords :
            l2.append(word)
            tf_idf_score += getScore(word, word_dict)

    # 根据TextRank()算法提取出关键词
    TextRank_keyword_list = analyse.textrank(word_list, topK=50, withWeight=True)

    # 去掉空语气词并保留其语气词词频，并计算其情感得分
    l3 = []
    TextRank_score = 0
    for word in TextRank_keyword_list:
        if not word[0].isdigit() and not word[0].isspace() and word[0] not in stopwords:
            l3.append(word)
            TextRank_score += getScore(word, word_dict)
            
    print(tf_idf_score, TextRank_score)
```
